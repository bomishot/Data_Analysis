# BeautifulSoup 이용

```python
import requests
```

```python
req = requests.get("https://m.blog.naver.com/luvpolo/221300832833") # url 주소
# req.text
```

f12버튼을 누르면 개발자 도구가 나온다! html이 어떻게 구성되었는지 확인 가능  -> 확인을 통해 우리가 필요한 것만 크롤링 가능  
req.text를 통해 이 html불러온다  
이것을 아름답게 처리해주는 것이 beautifulsoup  
req.text를 통해 좀 더러운 것들이 req.content를 통해 조금의 정리가 들어가게 나온다!

### req.content

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(req.content, 'html.parser')
# print(soup) -> html태그 길게 나오는데 너무 길어서 ..
```

```python
#soup.find('span') # span중에 가장 처음에 있는 span을 찾는다
span_list = soup.find('tbody').find_all('span') # 모든 span값들 찾아줌
span_list
```

## 쓸데없는거빼고 텍스트화해보자!

```python
soup.find('tbody').find_all('span')[0]
```

```python
soup.find('tbody').find_all('span')[0].text
```

```python
span_list2=list()
for span in span_list:
    span_list2.append(span.text)

print(span_list2)
```

## 쓸데없는 제목 빼고 !! 쓸데없는 숫자도 빼자!

```python
span_list2[0]
```

```python
del span_list2[0]
```

```python
span_list2
```

## 문자열이 세자리보다 작을때 날리자!

```python
span_list3=list()
for span in span_list2:
    if len(span)>3:
        span_list3.append(span)
span_list3
```
